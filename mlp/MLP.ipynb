```python
import pandas
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("/kaggle/input/clean12345/clean.csv")
```

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
```

```python
y_train=tf.constant(train_df['price'])/1000000
X_train=train_df.drop(columns=['price'])

y_test=tf.constant(test_df['price'])/1000000
X_test=test_df.drop(columns=['price'])
```

```python
scaler_minmax = MinMaxScaler()
X_train = scaler_minmax.fit_transform(X_train)

scaler_minmax = MinMaxScaler()
X_test = scaler_minmax.fit_transform(X_test)

X_train=tf.constant(train_df)
X_test=tf.constant(test_df)
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(16,)),
    Dense(128, activation='relu'),
    Dense(1)  
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32, verbose=1, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)
```
